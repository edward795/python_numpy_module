<h2>Plotting a displot</h2>

```python
import matplotlib.pyplot as plt 
import seaborn as sns

sns.distplot([1,2,3,4,5])
plt.show()
```

<h2>Plotting a displot with histogram</h2>

```python
import matplotlib.pyplot as plt 
import seaborn as sns

sns.distplot([1,2,3,4,5],hist=False)
plt.show()
```
